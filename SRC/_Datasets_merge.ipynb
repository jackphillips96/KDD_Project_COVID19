```python
# Import the `pandas` library as `pd`. And load 'stay at home' data set  as df1. after loading data set we renamed 'Entety' to loaction as we need to merge data sets on common values. 
import pandas as pd
df1 = pd.read_csv('stay-at-home-covid.csv') #path may have to be altered
df1=df1.rename(columns={'Entity':'location'})
df1
```

```python
#loading 'owid-covid-data' as df2 and renaming 'iso_code' to 'Code' to enable merging data sets  on common values.
df2 = pd.read_csv('owid-covid-data.csv')
df2=df2.rename(columns={'iso_code':'Code'})
df2
```

```python
#utilize merge finction to merge to data sets on common values such as 'location', 'Date', 'Code'
covid_data=pd.merge(df1, df2, on=['location', 'Date', 'Code'])
covid_data
```
